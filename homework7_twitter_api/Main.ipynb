```python
target_users = ["BBC", "CBS", "CNN", "Fox"]


df = pd.DataFrame(columns= ["Counter", "User", "News", "Negative", "Positive", "Neutral" , "Compound"])




for i in target_users:
    counter = 0
    target_user = i
    
    for x in range(5):

        public_tweets = api.user_timeline(target_user)
        
        
        for j in public_tweets:

            text = j["text"]
            
            x = analyzer.polarity_scores(text)
            
            Negative = x["neg"]
            Positive  = x["pos"]
            Neutral = x["neu"]
            Compound = x["compound"]
            
            counter +=1
            
            
            df2 = pd.DataFrame([[counter, i, text, Negative, Positive, Neutral , Compound]], 
                               columns= ["counter","user", "news","Negative","Positive","Neutral","Compound"])
            
            
            
            df = df.append(df2)
            
            
            
            
```
